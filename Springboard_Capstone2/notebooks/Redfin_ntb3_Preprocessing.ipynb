# Preprocessing - Redfin Data

```python
#load modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#from sklearn.model_selection import train_test_split
#from sklearn.preprocessing import StandardScaler
import category_encoders as ce
```

```python
# load data
df = pd.read_csv('../data/interim/redfin_data_to_process.csv', index_col=0)
df.head()
```

```python
df.info()
```

```python
# OneHotEncoding the city column
X_city = df['city']
encoder = ce.OneHotEncoder(cols=['city'])
X_clean_city = encoder.fit_transform(X_city)
encoder.get_feature_names()
X_clean_city.value_counts()
```

```python
df['city'].value_counts()
```

```python
X_clean_city = X_clean_city.rename(columns={'city_1': 'city_LasVegas', 'city_2':'city_Henderson', 'city_3':'city_BlueDiamond', 'city_4':'city_BoulderCity', 'city_5':'city_NorthLasVegas', 'city_6':'city_Enterprise'})
X_clean_city.head()
```

```python
df = pd.concat([df, X_clean_city], axis=1)
df.head()
```

```python
#checking to make sure I labelled cities correctly
df[df['city'] == 'Blue Diamond']
```

```python
# df_final
#I'm going to drop the column for 'city_LasVegas' because it is unnecessary
drop_columns = ['address', 'city', 'price_per_sqft', 'property_type', 'city_LasVegas']
df_final = df.drop(columns=drop_columns)
df_final.head()
```

```python
df_final.to_csv('../data/processed/redfin_data_processed.csv')
```
