```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
#concatenating downloaded csv files from Redfin
#Redfin has download limit of 350 records
#to get around this I used the following criteria in my search:
    #searched by zipcode, houses sold in the last 3 months
df = pd.concat(map(pd.read_csv, glob.glob('redfin_data/*.csv')))
```

```python
#get a general sense of the data
df.info()
```

```python
df.head()
```

```python
#drop the columns that are not useful or interpretable
df = df.drop(columns=['SOLD DATE', 'SALE TYPE', 'STATUS', 'NEXT OPEN HOUSE START TIME', 'NEXT OPEN HOUSE END TIME', 'URL (SEE http://www.redfin.com/buy-a-home/comparative-market-analysis FOR INFO ON PRICING)', 'SOURCE', 'MLS#', 'FAVORITE', 'INTERESTED'])
```

```python
df.head()
```

```python
#look at the unique values in the 'Property Type' column, will break these out into columns later
print(df['PROPERTY TYPE'].unique())
```

```python
df['PROPERTY TYPE'].value_counts()
```

```python
#drop if 'Property Type' == 'Multi-Family (2-4 Unit)', 'Other', 'Vacant Land', 'Multi-Family (5+ Unit)'
df = df[df['PROPERTY TYPE'] != 'Other']
df = df[df['PROPERTY TYPE'] != 'Multi-Family (2-4 Unit)']
df = df[df['PROPERTY TYPE'] != 'Vacant Land']
df = df[df['PROPERTY TYPE'] != 'Multi-Family (5+ Unit)']

df['PROPERTY TYPE'].value_counts()
```

```python
#drop the duplicate rows
df.drop_duplicates(subset=['ADDRESS'], inplace=True)
df.info()
```

```python
#total nans in each column of the df, will use this information to determine how to address the nans
df.isna().sum()
```

```python
#dealing with nans
#dropping the 1 row with missing address
df.dropna(subset=['ADDRESS'], inplace=True)
```

```python
#probably won't use the location, because zipcode is also representative of the location, but going to keep for now
locations_list = df.LOCATION.unique().tolist()
print(locations_list)
```

```python
#replace location nans with 'Other'
df['LOCATION'].fillna(value='Other', inplace=True)
```

```python
#evaluating progress
df.isna().sum()
```

```python
df[df['SQUARE FEET'].isna()]
```

```python
#the rows with nans in 'Sqaure Feet' column also have a lot of other missing data, so I've decided to drop these records altogether
df.dropna(subset=['SQUARE FEET'], inplace=True)
```

```python
df.isna().sum()
```

```python
#I'm also dropping all rows with nans in the 'BEDS' or 'BATHS' column
df.dropna(subset=['BEDS', 'BATHS'], inplace=True)
df.BEDS.value_counts()
```

```python
df.BATHS.value_counts()
```

```python
df.isna().sum()
```

```python
df['PROPERTY TYPE'].value_counts()
```

```python
df['CITY'].value_counts()
```

```python
df['CITY'].replace(to_replace=['LAS VEGAS', 'HENDERSON', 'Mt Charleston'], value=['Las Vegas', 'Henderson', 'Mount Charleston'], inplace=True)
cities_keep = ['Las Vegas', 'Henderson', 'North Las Vegas', 'Boulder City', 'Mount Charleston', 'Blue Diamond', 'Enterprise']
df = df[df['CITY'].isin(cities_keep)]
df['CITY'].value_counts()
```

```python
city_price_means = df.groupby(by='CITY')['PRICE'].mean()
city_price_means.head()
```

```python
city_price_means.sort_values(ascending=False).plot(kind='barh', figsize=(5,5), title='Average property price by City')
plt.xlabel('Price ($)')
plt.show()
```

```python
zipcode_price_means = df.groupby(by='ZIP OR POSTAL CODE')['PRICE'].mean()
zipcode_price_means.head()
zipcode_price_means.sort_values(ascending=False).plot(kind='barh', figsize=(10,20), title='Average property price by Zipcode')
plt.xlabel('Price ($)')
plt.show()
```

The zipcode 89158 is actually on the highend portion Las Vegas Strip, so it makes sense that the housing prices would be much higher here.

```python
# before I decide how to handle the remaining nans, I'm plotting histograms of all the numeric features
df.hist(figsize=(15,10))
plt.subplots_adjust(hspace=0.5)
```

I can ignore zipcode here, while it appears to be a numeric feature, it is really categorical. I will look at distribution of price per zipcode a little bit further down

Some interesting insights here:
 - number of beds has a fairly normal distribution
 - it looks like Baths, Square Feet, Lot Size, $/Square Feet, and HOA/Month might have some outliers
 - 2 of the columns with missing values are concerning because they clustered at the low end: Lot Size & HOA/Month

```python
#renaming columns
df = df.rename(str.lower, axis='columns')
df = df.rename(columns={'property type': 'property_type', 'state or province': 'state', 'zip or postal code': 'zipcode', 'square feet': 'square_feet', 'lot size': 'lot_size', 'year built': 'year_built', 'days on market': 'days_on_market', '$/square feet': 'price_per_sqft', 'hoa/month': 'hoa_per_mon'})
df.info()
```

```python
df.describe().T
```

# Lot Size - drop nans and outliers

```python
df.lot_size.hist(bins=30)
plt.xlabel('lot_size')
plt.ylabel('count')
plt.title('Distribution of lot size')
plt.show()
```

```python
df['lot_size'].describe().T
```

```python
df['lot_size'].median()
```

```python
df.info()
```

```python
#dropping all lot sizes larger than 87120 sq_ft (2 acres) and nans
df = df[df.lot_size <= 87120]
```

```python
df.lot_size.hist(bins=30)
plt.xlabel('lot_size')
plt.ylabel('count')
plt.title('Distribution of lot size after dropping properties over 2 acres')
plt.show()
```

```python
df.info()
```

# Days On Market - replace nans

```python
df['days_on_market'].describe().T
```

```python
df['days_on_market'].median()
```

```python
df['days_on_market'] = df['days_on_market'].fillna(value = df['days_on_market'].mean())
```

```python
df['days_on_market'].describe().T
```

# HOA per month - replace nans

```python
df.hoa_per_mon.hist(bins=30)
plt.xlabel('hoa_per_month')
plt.ylabel('count')
plt.title('Distribution of hoa per month')
plt.show()
```

```python
df[df['hoa_per_mon'] > 1000]
```

```python
df['hoa_per_mon'].describe().T
```

```python
df['hoa_per_mon'].median()
```

```python
# since it looks like there might be quite a few outliers in this column, I'm going to use median to fillna
df['hoa_per_mon'] = df['hoa_per_mon'].fillna(value = df['hoa_per_mon'].median())
df['hoa_per_mon'].describe().T
```

```python
df.hoa_per_mon.hist(bins=30)
plt.xlabel('hoa_per_month')
plt.ylabel('count')
plt.title('Distribution of hoa per month after filling nans')
plt.show()
```
